```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_subplot(data, ax=None):
    if ax is None:
        ax = plt.gca()
    plot = ax.plot(data)
    return plot
```

```python
data = np.random.randn(2,2)
another_data = np.random.randn(2,2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(10,5))
create_subplot(data, ax1)
create_subplot(another_data, ax2)
plt.show()
```
